```python
import pandas as pd
from sodapy import Socrata
import requests as rq
import json
```

### Metadata ###  
Raadpleeg de discovery API van Socrata voor ophalen van metadata opendata.rdw.nl.   

```python
metadata = rq.get('http://api.eu.socrata.com/api/catalog/v1?domains=opendata.rdw.nl')
metadata_json = json.loads(metadata.text)
```

```python
#Create name for all Type goedkeuring API
for result in metadata_json['results']:
    #print(result['resource']['description'])
    if (result['resource']['name'] =="Open Data RDW: EEG Voertuigtypegoedkeuring"):
        entlst = [("Open Data RDW: " + name[4:]) for name in result['resource']['columns_name'] if name[:3] == 'API']
entlst.append("Open Data RDW: EEG Voertuigtypegoedkeuring")
```

```python
typegoedkeuring = [(ent,result['resource']['columns_name']) for result in metadata_json['results'] for ent in entlst if result['resource']['name'].lower() == ent.lower() ]
```

```python
for line in typegoedkeuring:
    print(line[0])
    for attr in line[1]:
        print('\t',attr)
```

### Linked data ###  
Every occurence in typegoedkeuring array is an owl:Class with prefLabel de value in the array.  
The identifying URI looks like t99999. Where 99999 starts from 1 and is increased with 1.  
Every entry in the second column of the array is a quality (kwaliteit) of the entry in the first column.  
This means it is a member of de kwaliteit class and there is an relation heeftKwaliteit between  
the first column and the second column entries.

```python
#prefix URI's
base = '<http://www.me-ds.nl/voc/voertuig#'
owl = '<http://www.w3.org/2002/07/owl#'
rdf = '<http://www.w3.org/1999/02/22-rdf-syntax-ns#'
rdfs = '<http://www.w3.org/2000/01/rdf-schema#'
skos = '<http://www.w3.org/2004/02/skos/core#'
begrippenxl = 'https://www.begrippenxl.nl/rest/v1/rdw/data?uri=http%3A%2F%2Fwww.me-datastudio.nl%2Fvocabularies%2Fvoertuig%23'

#suffix
turtle = '&format=text/turtle'
jsonld = '&format=application/ld%2Bjson'

#basic Classes and Properties
kwaldef = '<http://www.me-ds.nl/voc/voertuig#Kwaliteit> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/2002/07/owl#Class> .'
kwallabel = '<http://www.me-ds.nl/voc/voertuig#Kwaliteit> <http://www.w3.org/2004/02/skos/core#prefLabel> "Kwaliteit"@nl  .'
heeftKwal = '<http://www.me-ds.nl/voc/voertuig#heeftKwaliteit> <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/2002/07/owl#ObjectProperty> .'
objprop = '<http://www.me-ds.nl/voc/voertuig#heeftKwaliteit> <http://www.w3.org/2000/01/rdf-schema#subPropertyOf> <http://www.w3.org/2002/07/owl#topObjectProperty> .'
subclassPO = ' <http://www.w3.org/2000/01/rdf-schema#subClassOf> <http://www.me-ds.nl/voc/voertuig#Kwaliteit> .'
subpropPO = ' <http://www.w3.org/2000/01/rdf-schema#subPropertyOf> <http://www.me-ds.nl/voc/voertuig#heeftKwaliteit> .'
objpropPO = ' <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/2002/07/owl#ObjectProperty> .'
subobjpropPO = ' <http://www.w3.org/2000/01/rdf-schema#subPropertyOf> <http://www.w3.org/2002/07/owl#topObjectProperty> .'
```

```python
f = open('typegoedkeuring.owl','w')
f.write(kwaldef+'\n')
f.write(kwallabel+'\n')
f.write(heeftKwal+'\n')
f.write(objprop+'\n')
i = 1
i_lst = []
i_root = 0
for line in typegoedkeuring:
    cluri = '<http://www.me-ds.nl/voc/voertuig#t' + str(i) + '>'
    cl =  cluri + ' <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/2002/07/owl#Class> .'
    cp =  cluri + ' <http://www.w3.org/2004/02/skos/core#prefLabel> "'+ line[0][15:] + '"@nl  .'
    f.write(cl+'\n')
    f.write(cp+'\n')
    
    if line[0] != 'Open Data RDW: EEG Voertuigtypegoedkeuring':
        i_lst.append(i)
    else:
        i_root = i
 
    i=i+1
    j=100
    for attr in line[1]:
        if attr[:3]!='API':
            acuri = '<http://www.me-ds.nl/voc/voertuig#t' + str(i) + str(j) + '>'
            opuri = '<http://www.me-ds.nl/voc/voertuig#heeftT' + str(i) + str(j) + '>'
            ac = acuri + ' <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> <http://www.w3.org/2002/07/owl#Class> .'
            ap = acuri + ' <http://www.w3.org/2004/02/skos/core#prefLabel> "'+ attr + '"@nl  .'
            asc = acuri + subclassPO
            aop = opuri + subpropPO
            opd = opuri + ' <http://www.w3.org/2000/01/rdf-schema#domain> ' + cluri + ' .'
            opr = opuri + ' <http://www.w3.org/2000/01/rdf-schema#range> ' + acuri + ' .'
            f.write(ac+'\n')
            f.write(ap+'\n')
            f.write(asc+'\n')
            f.write(aop+'\n')
            f.write(opd+'\n')
            f.write(opr+'\n')
            j=j+1
            
#maak ObjectProperty onderdeelVan aan onder topObjectProperty            
opuri = '<http://www.me-ds.nl/voc/voertuig#onderdeelVan>'  
f.write(opuri + objpropPO + '\n')
f.write(opuri + subobjpropPO + '\n')
#maak API's onderdeel van EEG Voertuigtypegoedkeuring via objectproperty onderdeelVan domain en range
for val in i_lst:
    cluri = '<http://www.me-ds.nl/voc/voertuig#t' + str(val) + '>'
    rturi = '<http://www.me-ds.nl/voc/voertuig#t' + str(i_root) + '>'
    opr = opuri + ' <http://www.w3.org/2000/01/rdf-schema#range> ' + cluri + ' .'
    opd = opuri + ' <http://www.w3.org/2000/01/rdf-schema#domain> ' + rturi + ' .'
    f.write(opd+'\n')
    f.write(opr+'\n')
f.close()
```
